```python
logging.basicConfig(format='[%(asctime)s] %(message)s', level=logging.INFO)
```

```python
sentences = sv.load("Wiki_Sentences_SP")
```

```python
vocab = set()
filtered_phrases = set()
for s,sent in enumerate(sentences):
    total = len(sent)
    i = 0
    try:
        while i<total-1:
            if sent[i]=='[':
                i+=1
                phrase = list()
                while sent[i]!=']':
                    if sent[i][0].isupper() and sent[i][1:].islower():
                        phrase.append(sent[i])
                    i+=1
                if phrase:
                    filtered_phrases.add(tuple(phrase))
                    vocab.update(set(phrase))
            else:
                i+=1
    except:
        #logging.info("sentence: {} of size {}, overflowed at {} last word was: {}".format(s,total,i,sent[i-1]))
        pass
    if s%10000 == 0:
        logging.info("Done with sentence: {} vocab reached: {}".format(s,len(vocab)))
```

```python
sv.save(vocab,"filtered_phrase_vocab")
```
